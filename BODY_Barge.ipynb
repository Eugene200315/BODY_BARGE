```python
import pychrono as chrono
import pychrono.irrlicht as chronoirr
import math as m
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
df=pd.read_excel("body_response.xlsx")
arrayti=df['TIME'].tolist()
arrayfx=df['FX'].tolist()
arrayfy=df['FY'].tolist()
arraymz=df['MZ'].tolist()

rest_length = 0.6
spring_coef =600
damping_coef = 0
b_m=17.176


class MySpringForce(chrono.ForceFunctor):
    def __init__(self):
        super(MySpringForce, self).__init__()
    
    def __call__(             self,         #
                              time,         # current time
                              rest_length,  # undeformed length
                              length,       # current length
                              vel,          # current velocity (positive when extending)
                              link):         # back-pointer to associated link
                              
        force = -spring_coef * (length - rest_length) - damping_coef * vel
        return force
    


system = chrono.ChSystemNSC()
system.Set_G_acc(chrono.ChVectorD(0, 0, 0))

# Create the ground body with two visualization spheres
ground = chrono.ChBody()
system.AddBody(ground)
ground.SetIdentifier(-1)
ground.SetBodyFixed(True)
ground.SetCollide(False)

sph_1 = chrono.ChSphereShape()
sph_1.GetSphereGeometry().rad = 0.04
sph_1.Pos = chrono.ChVectorD(0.77, 0, 0)
ground.AddAsset(sph_1)

sph_2 = chrono.ChSphereShape()
sph_2.GetSphereGeometry().rad = 0.04
sph_2.Pos = chrono.ChVectorD(-0.77, 0, 0)
ground.AddAsset(sph_2)
POS=chrono.ChVectorD()
ROT=chrono.ChVectorD()


body_2 = chrono.ChBody()
system.AddBody(body_2)
body_2.SetPos(chrono.ChVectorD(0,0, 0))
body_2.SetIdentifier(1)
body_2.SetBodyFixed(False)
body_2.SetCollide(True)
body_2.SetMass(b_m)
body_2.SetInertiaXX(chrono.ChVectorD(0.001,0.001, 0.1072))

# Attach a visualization asset.
box_2 = chrono.ChBoxShape()
box_2.GetBoxGeometry().SetLengths(chrono.ChVectorD(0.3, 0.2, 0.58))
body_2.AddAsset(box_2)
col_2 = chrono.ChColorAsset()
col_2.SetColor(chrono.ChColor(0, 0, 0.6))
body_2.AddAsset(col_2)


# Create the spring between body_2 and ground. The spring end points are

force = MySpringForce()

spring_2 = chrono.ChLinkTSDA()
spring_2.Initialize(body_2, ground, True, chrono.ChVectorD(0,0 , 0), chrono.ChVectorD(0.75,0, 0), False, rest_length)
spring_2.RegisterForceFunctor(force)
system.AddLink(spring_2)

# Attach a visualization asset.
spring_2.AddAsset(col_2)
spring_2.AddAsset(chrono.ChPointPointSpring(0.05, 80, 15))
# Create the spring between body_1 and ground. The spring end points are
# specified in the body relative frames.
spring_1 = chrono.ChLinkTSDA()
spring_1.Initialize(body_2, ground, True, chrono.ChVectorD(0, 0, 0), chrono.ChVectorD(-0.75,0, 0), False, rest_length)
spring_1.SetSpringCoefficient(spring_coef)
spring_1.SetDampingCoefficient(damping_coef)
system.AddLink(spring_1)

# Attach a visualization asset.
col_1 = chrono.ChColorAsset()
col_1.SetColor(chrono.ChColor(0.6, 0, 0))
spring_1.AddAsset(col_1)
spring_1.AddAsset(chrono.ChPointPointSpring(0.05, 80, 15))
#arrayin=chrono.ChVectorD()
#arrayin= body_2.GetInertiaXX()
#print(arrayin)
#Add FORCE IN BODY


application = chronoirr.ChIrrApp(system, "ChLinkTSDA demo", chronoirr.dimension2du(800, 600))
application.AddTypicalLogo()
application.AddTypicalSky()
application.AddTypicalLights()
application.AddTypicalCamera(chronoirr.vector3df(0, 0, 2))

application.AssetBindAll()
application.AssetUpdateAll()

# Simulation loop

pos_y  = []
#arrayfo   =[]
#arraysp   =[]
thetha=[]
pos_x=[]
application.SetTimestep(0.002)
application.SetTryRealtime(True)
fra=0

while (application.GetDevice().run()):
    tim=round(system.GetChTime(),4)  
    body_2.Empty_forces_accumulators()  
    TZ= chrono.ChVectorD(0,0,arraymz[arrayti.index(tim)])    
    FORCE=chrono.ChVectorD(0*arrayfx[arrayti.index(tim)],arrayfy[arrayti.index(tim)],0)  #
    fra+=1
    POS=body_2.GetPos()
    ROT= body_2.GetRotAngle()
    #print((180/3.14)*ROT)
    r0=body_2.GetRotAxis()
    #print(r0.z)
    print(r0.z*(180/3.14)*ROT)
    pos_y.append(body_2.GetPos().y)
    pos_x.append(body_2.GetPos().x)
    body_2.Accumulate_torque(TZ,True) 
    thetha.append(r0.z*(180/3.14)*ROT)
    body_2.Accumulate_force(FORCE,POS,False) 
    wp=body_2.TransformPointParentToLocal(POS)
    #wpp=np.array(wp)
    #print(wpp)
    application.BeginScene()
    application.DrawAll()
    application.DoStep()
    application.EndScene()
    if tim>=arrayti[-1]:
          application.GetDevice().closeDevice()       
    
 # Use matplotlib to make two plots when simulation ended:
fig, (ax1, ax2,ax3) = plt.subplots(3, sharex = True)

ax1.plot( arrayti,pos_y,'r--' )
ax1.set(ylabel='HEAVE [m]',xlabel='Time [t]')
ax1.grid()

ax2.plot( arrayti,thetha , 'b--')
ax2.set(ylabel='ROLL [deg]',xlabel='Time [t]')
ax2.grid()


ax3.plot( arrayti,pos_x , 'g--')
ax3.set(ylabel='Sway [m]',xlabel='Time [t]')
ax3.grid()
```
